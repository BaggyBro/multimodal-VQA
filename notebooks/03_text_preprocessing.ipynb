```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('../data/data_train.csv')
print(train_df.columns)
print(train_df['answer'].head())
```

```python
# stripping whitespaces
train_df['clean_question'] = train_df['question'].str.strip().str.lower()

# missing/null questions
missing_count = train_df['question'].isnull().sum()
print(f"Missing questions: {missing_count}")

train_df = train_df.dropna(subset='clean_question')
```

```python
train_df['question_length'] = train_df['clean_question'].apply(len)

plt.hist(train_df['question_length'], bins=30)
plt.title('Question Length Distribution')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

```python
train_df.to_csv('../data/data_train_cleaned.csv', index=False)
```
