```python
import pandas as pd
from autogluon.multimodal import MultiModalPredictor
from sklearn.model_selection import train_test_split
import os
```

```python
train_df = pd.read_csv('../data/data_train_cleaned.csv')

# add actual image file paths
image_dir='../data/images_resized/'
train_df['image_path'] = train_df['image_id'].apply(lambda x : os.path.join(image_dir,f"{x}.png"))

train_split, val_split = train_test_split(train_df, test_size=0.2, random_state=42)

train_split.to_csv('../data/data_train_final.csv', index=False)
val_split.to_csv('../data/data_val_cleaned.csv', index=False)
```

```python
train_df.head()
```

```python
predictor = MultiModalPredictor(label="answer")
predictor.fit(train_split, tuning_data=val_split, time_limit=3600)
```
